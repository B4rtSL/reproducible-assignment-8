# Breaking Bad

<b>Breaking Bad</b> is an American crime drama television series created and produced by <i>Vince Gilligan</i> for AMC. Set and filmed in Albuquerque, New Mexico, the series follows <b>Walter White</b> (<i>Bryan Cranston</i>), an underpaid, dispirited high-school chemistry teacher struggling with a recent diagnosis of stage-three lung cancer. White turns to a life of crime and partners with a former student, <b>Jesse Pinkman</b> (<i>Aaron Paul</i>), to produce and distribute methamphetamine to secure his family's financial future before he dies, while navigating the dangers of the criminal underworld. Breaking Bad premiered on AMC on January 20, 2008, and concluded on September 29, 2013, after five seasons consisting of 62 episodes.

## Graphs

### Viewership by season

```python
import pandas as pd
import matplotlib.pyplot as plt

seasons = ['1', '2', '3', '4', '5']

avgViews = [1.23, 1.3, 1.52, 1.87, 4.32]

plt.plot(seasons, avgViews, marker='o', linestyle='-', label='First Aired Views', color = 'black')
plt.xlabel('Season')
plt.ylabel('Views (in millions)')
plt.title('Average Viewership Comparison by Season')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Viewership by time of airing the episode

```python
from datetime import datetime

# original dates from Wikipedia
firstDates = ['January 20, 2008', 'March 8, 2009', 'March 21, 2010', 'July 17, 2011', 'July 15, 2012', 'August 11, 2013']
lastDates = ['March 9, 2008', 'May 31, 2009', 'June 13, 2010', 'October 9, 2011', 'September 2, 2012', 'September 29, 2013']

firstDates = [datetime.strptime(date, '%B %d, %Y') for date in firstDates]
lastDates = [datetime.strptime(date, '%B %d, %Y') for date in lastDates]

firstViews = [1.41, 1.66, 1.95, 2.58, 2.93, 5.92]
lastViews = [1.5, 1.5, 1.56, 1.9, 2.78, 10.28]

plt.plot(firstDates, firstViews, marker='o', linestyle='-', label='First Aired Views', color='blue')
plt.plot(lastDates, lastViews, marker='s', linestyle='--', label='Last Aired Views', color = 'red')
plt.xlabel('Date')
plt.ylabel('Views (in millions)')
plt.title('Viewership Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
mean_last_views = sum(lastViews) / len(lastViews)

print("Mean viewership:", mean_last_views)
```
